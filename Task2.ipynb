```python
mark=int(input("Enter the mark:"))
if 90<mark<=100:
    print("Grade A with Distrinction")
elif 80<mark<=90:
    print("Grade B with First Class")
elif 70<mark<=80:
    print("Grade C with Second Class")
elif 60<mark<=70:
    print("Grade D with Third Class")
else:
    print("Better Luck Next Time")
```
